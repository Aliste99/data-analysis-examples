```python
import pandas as pd
```

```python
df = pd.read_csv('data/Minjust2018.csv', 
                 header=None, 
                 usecols=[4, 27, 28, 33])

df.fillna('', inplace=True)
df.rename(columns={4: 'name', 
                   27: 'director', 
                   28: 'activity', 
                   33: 'founders'},
              inplace = True)
```

```python
df['new_column'] = 0
df
```

```python
df['director_surname'] = df['director'].str.split().str[1]
```

```python
def count_letters(words):
    return pd.Series([len(word) for word in words])
```

```python
df.loc[count_letters(df['director']) == count_letters(df['name'])]
```
